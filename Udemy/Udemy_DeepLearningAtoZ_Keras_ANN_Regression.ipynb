Artificial Neural Networks - Regression

```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

1) Data Preprocessing

```python
# Dataset source: https://archive.ics.uci.edu/ml/datasets/combined+cycle+power+plant

df = pd.read_excel("Folds5x2_pp.xlsx")
```

```python
# Check NaN

df.info()
```

```python
X = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
y
```

```python
# Splitting the dataset into training and test set

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state =1)
```

2) Building the ANN

```python
ann = tf.keras.models.Sequential()
```

```python
ann.add(tf.keras.layers.Dense(units=6, activation="relu"))
ann.add(tf.keras.layers.Dense(units=6, activation="relu"))

ann.add(tf.keras.layers.Dense(units=1)) # For regression problems -> No activation function
```

```python
ann.compile(optimizer='adam', loss='mean_squared_error')
```

3) Training the ANN

```python
ann.fit(X_train, y_train, batch_size = 32, epochs=100)
```

```python
# Predicting the results

y_pred = ann.predict(X_test)
np.set_printoptions(precision=2)
y_compare = np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)), 1)
y_compare
```

```python
# Evaluation 

from sklearn.metrics import mean_squared_error

print("MSE (test): ", mean_squared_error(y_test, y_pred))

print("MSE (test): ", ann.evaluate(X_test, y_test, batch_size=32))
```
