```
import numpy as np
import pandas as pd
import tensorflow as tf
```

```
tf.__version__
```

**1) Data Preprocessing**

```
df = pd.read_csv("Churn_Modelling.csv")
df.head()
```

```
X = df.iloc[:,4:-1].values # array
y = df.iloc[:,-1].values # array

X_df = df.iloc[:,4:-1]
y_df = df.iloc[:,-1]
```

```
# a) Missinig values
df.info()

# No NaNs
```

```
X_df.head()
```

```
# b) Encoding categorical features

# Label encoding "Gender"
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X[:,1] = le.fit_transform(X[:,1])

X
```

```
# One-Hot-Encoding "Geography"

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct = ColumnTransformer(transformers=[("encoder", OneHotEncoder(),[0])], remainder="passthrough")
X = np.array(ct.fit_transform(X))
X
```

```
X[1]
```

```
# c) Splitting dataset into the training and test set

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```
# d) Feature Scaling

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

**2) Building an ANN**

a) Initializing the ANN

```
ann = tf.keras.models.Sequential()
```

b) Adding the input layer and the first hidden layer

```
ann.add(tf.keras.layers.Dense(units=6, activation='relu')) # units = 6 -> number of hidden neurons
```

c) Adding the second hidden layer

```
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

d) Adding the output layer

```
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

**3) Training the ANN**

a) Compiling the ANN

```
ann.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy']) # not binary problem: categorical_crossentropy and activation function = softmax
```

b) Training the ANN on the training set

```
ann.fit(X_train, y_train, batch_size = 32, epochs=100)
```

4) Making the prediction and evaluating the model

a) Predicting the result of a single observation

```
# Without "Credit score"!
ann.predict(sc.transform([[1.0, 1.0, 1.0, 1, 40, 3, 60000, 2, 1, 1, 50000]])) # Always in: [[]]-> 2D array
```

```
ann.predict(sc.transform([[1.0, 1.0, 1.0, 1, 40, 3, 60000, 2, 1, 1, 50000]])) >0.5
```

b) Predicting the test set results

```
y_pred = ann.predict(X_test)
y_pred = (y_pred > 0.5)
y_compare = np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)), 1)
y_compare
```

c) Making the Confusion Matrix

```
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
